```python
import os
from pyaxmlparser import APK
from keras.models import load_model
from tensorflow.keras.models import load_model

import numpy as np
```

```python
def process(arr1):
  res = []
  for i in arr1:
    temp = i.split('.')
    if 'permission' in temp:
      ind = temp.index('permission')
      res.append(".".join(temp[ind:]))
  return res
```

```python
from keras_radam import RAdam

predictionModel = load_model('setting/model.hdf5')
```

```python
#path-check
setPerms = []
f1 = open('list/PermissionListForPrediction.txt')
temp = f1.read()
setPerms = temp[2:-2].split("', '")
```

```python
#path-check
# apptest = "benign/QR code reader&QR code Scanner_3.6.1_Apkpure.apk"
apptest = "Untitled Folder/Android Malware/0038f450d7f1df75bf5890cf22299b0c99cc0bea8d66e6d25528cb01992a436b"

```

```python
a = APK(apptest)
perms = process(a.get_permissions())
print(perms)
```

```python
predictionList = [0] * len(setPerms)
for i in perms:
    if i in setPerms:
        predictionList[setPerms.index(i)] = 1
```

```python
predictionModel.predict(np.array(predictionList, ndmin=3))
```

```python
threshold = 0.5
prediction = predictionModel.predict(np.array(predictionList, ndmin=3))
predicted_value = "malware" if prediction >= threshold else "benign"
print(predicted_value)
```
